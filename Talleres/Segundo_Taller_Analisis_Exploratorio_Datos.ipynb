**1. Defina que es el análisis exploratorio de datos y explique su importancia en el
proceso de análisis de datos**

El análisis exploratorio de datos (EDA) es una técnica utilizada en el proceso de análisis de datos para comprender la naturaleza y las características de los datos que se están analizando. El EDA implica la realización de diversas técnicas de visualización de datos y cálculos estadísticos básicos para identificar patrones, tendencias, valores atípicos y otras características de los datos.

La importancia del EDA radica en su capacidad para identificar posibles problemas con los datos, como valores faltantes o valores atípicos, que pueden afectar los resultados del análisis posterior. Además, el EDA puede ayudar a identificar patrones y tendencias en los datos que pueden no ser evidentes de otra manera, lo que puede llevar a la identificación de nuevas relaciones entre variables o a la generación de nuevas hipótesis para investigar.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('exams_1.csv')

print(df.describe())
```

```python
sns.histplot(data=df, x="math score", binwidth=5, binrange=(0, 100), stat="count")

plt.xticks(range(0, 101, 5))
plt.yticks(range(0, 150, 10))
plt.xlabel("Puntaje de matemáticas")
plt.ylabel("Numero de estudiantes")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x="test preparation course", y="overall average", data=df)
plt.xlabel("Curso de preparación para exámenes")
plt.ylabel("Puntaje global")
plt.title("Diagrama de relacion en la realización del curso vs el puntaje global obtenido")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x="gender", y="overall average", data=df)
plt.xlabel("Género")
plt.ylabel("Puntaje de global")
plt.title("Diagrama de genero vs puntaje global")
plt.show()
```

**4.Tome la relación entre dos variables y explore como esta puede ayudar a explicar
los datos.**

```python

fig, ax = plt.subplots(figsize=(10, 5))
sns.scatterplot(x='parental level of education', y='overall average', data=df, ax=ax)
ax.set_xlim(left=-0.5, right=4.5)
plt.xlabel("Nivel de educación de los padres")
plt.ylabel("Puntaje de global")
plt.title("Relacion de nivel de educacion de los padres vs puntaje global de los estudiantes",)
plt.show()


```

```python
math_scores = df[['gender', 'math score']]
reading_scores = df[['gender', 'reading score']]
writing_scores = df[['gender', 'writing score']]

mean_math_scores = math_scores.groupby('gender').mean()
mean_reading_scores = reading_scores.groupby('gender').mean()
mean_writing_scores = writing_scores.groupby('gender').mean()

fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(12, 5))
mean_math_scores.plot(kind='bar', ax=ax[0], legend=None, title='Puntajes de Matemáticas por Género')
mean_reading_scores.plot(kind='bar', ax=ax[1], legend=None, title='Puntajes de Lectura por Género')
mean_writing_scores.plot(kind='bar', ax=ax[2], legend=None, title='Puntajes de Escritura por Género')

ax[0].set_xlabel('Género')
ax[1].set_xlabel('Género')
ax[2].set_xlabel('Género')
ax[0].set_ylabel('Puntaje Promedio')
plt.show()
```
